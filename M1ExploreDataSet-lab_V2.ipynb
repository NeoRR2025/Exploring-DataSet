Print the mean age of the survey participants.


```python
def age_to_numeric(age):
    if isinstance(age, str):
        if "Under 18" in age:
            return 17  # If age is under 18, we assign 17 as the typical age
        elif "18-24" in age:
            return 21  # Midpoint of 18-24
        elif "25-34" in age:
            return 29  # Midpoint of 25-34
        elif "35-44" in age:
            return 39  # Midpoint of 35-44
        elif "45-54" in age:
            return 49  # Midpoint of 45-54
        elif "55-64" in age:
            return 59  # Midpoint of 55-64
        elif "65 or older" in age:
            return 70  # An assumed average for 65 or older
    return None  # For missing or undefined age data

# Apply the conversion function to the 'Age' column
df['Age_numeric'] = df['Age'].apply(age_to_numeric)

# Calculate and print the mean age
mean_age = df['Age_numeric'].mean()

print(f"The mean age of the survey participants is: {mean_age}")
```

The dataset is the result of a world wide survey. Print how many unique countries are there in the Country column.


```python
# Print the number of unique countries in the 'Country' column
unique_countries = df['Country'].nunique()
print(f'Number of unique countries: {unique_countries}')
```
